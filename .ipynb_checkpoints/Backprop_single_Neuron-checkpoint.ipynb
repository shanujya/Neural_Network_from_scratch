```python
# initial params
weights = np.array([2.0,-3.0,4.0])
inputs = np.array([1.0,-1.0,1.0])
bias = 2.0
target_output = 0.0
learning_rate = 0.001
```

```python
def ReLu(x):
    return np.maximum(x,0)

def relu_derivative(x):
    return np.where(x>0,1.0,0.0)

for iteration in range(200):
    # Forward Pass
    linear_output = np.dot(weights,inputs)+bias
    output = ReLu(linear_output)
    loss = (output-target_output)**2
    
    # Backward Pass
    dloss_doutput = 2*(output-target_output)
    doutput_dlinear = relu_derivative(output)
    dlinear_dweight = inputs
    dlinear_dbias = 1.0
    
    dloss_dlinear = dloss_doutput*doutput_dlinear
    dloss_dweight = dloss_dlinear*dlinear_dweight
    dloss_dbias = dloss_dlinear*dlinear_dbias
    
    # Update weight and biases
    weights -= learning_rate*dloss_dweight
    bias -= learning_rate*dloss_dbias
    
    # Print the loss for this iteration
    print(f"Iterations:{iteration+1},Loss:{loss}")
          
print("Final Weights",weights)
print("Final BIas",bias)
```
